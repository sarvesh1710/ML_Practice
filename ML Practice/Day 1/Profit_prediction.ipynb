### Profit Prediction

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('State', axis=1, inplace=True)
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.pairplot(df)
```

```python
df.drop('Administration', axis=1, inplace=True)
```

```python
x = df.drop('Profit', axis=1)
y = df['Profit']
```

```python
from sklearn.linear_model import LinearRegression

# create the model
model = LinearRegression()

# train the model
model.fit(x, y)
```

```python
model.score(x, y)
```

```python
import pickle

with open("profit_prediction_model.pkl", "wb") as file:
    pickle.dump(model, file)
```
