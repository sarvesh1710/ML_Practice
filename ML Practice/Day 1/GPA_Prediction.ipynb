```python
plt.scatter(df['SAT Score'],df['GPA'])
plt.xlabel('SAT Score')
plt.ylabel('GAP')
plt.title("SAT Score vs GPA")
plt.legend()
```

```python
x = df.drop("GPA",axis = 1)
y = df['GPA']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x,y)
```

```python
model.score(x,y)
```

```python

model.predict([[1200]])
```

```python
predictions = model.predict(x)

plt.scatter(df['SAT Score'],df['GPA'],label = "actual values")
plt.scatter(df['SAT Score'],predictions,label = "Predicted values")
plt.plot(df['SAT Score'],predictions,label = "Best fit regression line")
plt.xlabel('SAT Score')
plt.ylabel('GAP')
plt.title("SAT Score vs GPA")
plt.legend()
```
