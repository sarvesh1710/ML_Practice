```python
x = df.drop('test_result',axis = 1)
y = df['test_result']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=123456)
```

## Support vector machine

```python
from sklearn.svm import SVC

# input the parameters from the hyper parameter tunning
model = SVC(C=2, gamma='scale', kernel ='rbf')

model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1, 2], 
    'gamma': ['scale', 'auto']
}

grid_search_svm = GridSearchCV(model,parameters)
grid_search_svm.fit(x_test,y_test)
```

```python
grid_search_svm.cv_results_
```

```python
grid_search_svm.best_params_
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f"accuracy = {accuracy:.2f}, precision = {precision:.2f}, recall = {recall:.2f}, f1 = {f1:.2f}")
```

### KNN model

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors = 7, weights='uniform')

knn_model.fit(x_train,y_train)
```

```python
y_pred_knn = knn_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred_knn)
precision = precision_score(y_test, y_pred_knn)
recall = recall_score(y_test, y_pred_knn)
f1 = f1_score(y_test, y_pred_knn)
print(f"accuracy = {accuracy:.2f}, precision = {precision:.2f}, recall = {recall:.2f}, f1 = {f1:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_neighbors': range(3, 10),
    'weights': ['uniform', 'distance']
}

knn_grid_search = GridSearchCV(knn_model,parameters)

knn_grid_search.fit(x_test,y_test)
```

```python
knn_grid_search.cv_results_
```
