```python
# since the features like name, home.dest are add no value while predicting the survivalness of a person
# we will remove them from the dataset
df.drop(["home.dest","name"],axis = 1, inplace=True)
```

```python
# since there are 263 records missing in age column we will replace the missing values 
# with the mean value of the age column
```

```python
age_mean = df['age'].mean()
df['age']=df['age'].fillna(age_mean)
```

```python
# since only two values are missing and no of records are also small (1309), we will replace 
# the NA records in embarked columns with its mode value
mode = df['embarked'].mode()[0]
df['embarked'] = df['embarked'].fillna(mode)
```

```python
df['embarked'].isna().sum()
```

```python
# since there is only one value is missing in fare, we will replace it with mean value
fare_mean = df['fare'].mean()
df['fare'] = df['fare'].fillna(fare_mean)
```

## Data cleansing

```python
df.rename({"sex":"gender"},axis = 1,inplace = True)
```

```python
# convert the non numerical columnn to numerical ones
from sklearn.preprocessing import LabelEncoder

encoder_gender = LabelEncoder()

df['gender'] = LabelEncoder().fit_transform(df['gender'])
df['ticket'] = LabelEncoder().fit_transform(df['ticket'])
df['embarked'] = LabelEncoder().fit_transform(df['embarked'])
```

### Split the data

```python
# create x and y
x = df[['pclass','gender','fare','embarked']]

y = df['survived']
```

```python
df['pclass'].unique()
```

```python
df['gender'].unique()
```

```python
df['embarked'].unique()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=123456)
```

### model building

```python
from sklearn.linear_model import LogisticRegressionCV

def create_logistic_regression():

    model = LogisticRegressionCV()

    model.fit(x_train,y_train)

    return model
```

```python
from sklearn.naive_bayes import GaussianNB

def create_naive_bayes_model():

    model = GaussianNB()

    model.fit(x_train,y_train)

    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier

def create_KNN_model5():

    model = KNeighborsClassifier(n_neighbors=5)

    model.fit(x_train,y_train)

    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier

def create_KNN_model4():

    model = KNeighborsClassifier(n_neighbors=4)

    model.fit(x_train,y_train)

    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier

def create_KNN_model3():

    model = KNeighborsClassifier(n_neighbors=3)

    model.fit(x_train,y_train)

    return model
```

### model evalation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaulate_model(model):
    y_pred = model.predict(x_test)

    accuracy = accuracy_score(y_test,y_pred)

    precision = precision_score(y_test,y_pred)

    recall  = recall_score(y_test,y_pred)

    f1 = f1_score(y_test,y_pred)

    return f"{accuracy:.2f}", f"{precision:.2f}", f"{recall:.2f}", f"{f1:.2f}"
```

```python
# list to keep al the models

models = []

models.append({"name": "Logistic Regression", "model" : create_logistic_regression()})
models.append({"name": "Naive bayes", "model" : create_naive_bayes_model()})
models.append({"name": "KNN5", "model" : create_KNN_model5()})
models.append({"name": "KNN3", "model" : create_KNN_model3()})
models.append({"name": "KNN4", "model" : create_KNN_model4()})
```

```python
# get the models evaluated

for model_info in models:
    accuracy,precision,recall,f1 = evaulate_model(model_info['model'])

    model_info['accuracy'] = accuracy
    model_info['precision'] = precision
    model_info['recall'] = recall
    model_info['f1'] = f1
```

```python
evaluation_df = pd.DataFrame(models,columns = ["name","accuracy","precision","recall","f1"])
evaluation_df
```
